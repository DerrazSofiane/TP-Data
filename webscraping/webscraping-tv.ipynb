```python
from bs4 import BeautifulSoup
import requests
import json
from subprocess import check_output
```

```python
url = "https://www.cdiscount.com/high-tech/televiseurs/tv-full-hd/l-1062608.html#_his_"
```

```python
response = requests.get(url)
print(type(response.content))
response.content
```

```python
content = BeautifulSoup(response.content)
print(type(content))
content
```

```python
produits = content.findAll("div", attrs={"class": "jsPrdBlocContainer"})
print(len(produits))
produits                         
```

```python
produit = produits[0]
produit
```

```python
text = produit.find("span", attrs={"class": "price"}).text
print(type(text))
text
```

```python
price = float(text.replace("€", "."))
price
```

```python
price = float(produit.find("span", attrs={"class": "price"}).text.replace("€", "."))

div = produit.find("div", attrs={"class": "prdtBILDetails"}).text
link = produit.find("a")
href = link["href"]

obj = {}
obj["price"] = price
obj["href"] = href


obj
```

```python
objs = []

for produit in produits : 
    price = float(produit.find("span", attrs={"class": "price"}).text.replace("€", "."))

    link = produit.find("a")
    href = link["href"]

    obj = {}
    obj["price"] = price
    obj["href"] = href

    objs.append(obj)
```

```python
print(len(objs))
objs
```

```python
objs.sort(key = lambda obj: obj["price"], reverse = True)
objs
```

```python
objs_top3 = objs[:3]
objs_top3
```

```python
with open("/temp.json", 'w') as outfile:
    json.dump(objs, outfile)
```
